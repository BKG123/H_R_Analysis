```python
df['salary'].isnull().sum()
```

```python
df_sal = pd.get_dummies(df['salary'],drop_first=True)
```

```python
df_sal.head()
```

```python
df=pd.concat([df,df_sal], axis=1)
```

```python
df.drop('salary',axis=1,inplace= True)
```

```python
df.head()
```

```python
df_dept = pd.get_dummies(df['Department'],drop_first=True)
```

```python
df_dept
```

```python
df=pd.concat([df,df_dept], axis=1)
```

```python
df.drop('Department',axis=1,inplace=True)
```

```python
df.head()
```

```python
sns.pairplot(df)
```

```python
y = df['left']
X = df.drop('left',axis=1)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,train_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
model.predict(X_test)
```

```python
model.score(X_test,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(max_depth=2, random_state=0)
clf.fit(X_train, y_train)
#print(clf.predict([[0, 0, 0, 0]]))
```

```python
clf.score(X_test,y_test)
```

```python
from sklearn.ensemble import AdaBoostClassifier
clf1 = AdaBoostClassifier(n_estimators=100, random_state=0)
clf1.fit(X_test, y_test)
print(clf1.predict(X_test))
clf1.score(X_test, y_test)
```

```python
from sklearn.metrics import classification_report
target_names = ['Left', 'Stayed']
print(classification_report(y_test, clf1.predict(X_test), target_names=target_names))
```
